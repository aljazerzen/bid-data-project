```python
import polars as pl
```

```python
df = pl.read_parquet('data/2022.parquet')
```

```python
bobaadr = pl.read_csv('data/pad/bobaadr.txt')
bobaadr.head()
```

```python
bobaadr.select(['sc5', 'scboro', 'hhnd', 'bin', 'physical_id', 'zipcode', 'stname']).write_parquet('./data/bobaadr.parquet')
bobaadr = pl.read_parquet('./data/bobaadr.parquet')
```

```python
bobaadr = (bobaadr
  .unique(subset=['sc5', 'hhnd'])
  .select([
    pl.col('sc5').alias('Street Code'),
    pl.col('scboro').cast(pl.Utf8).cast(pl.Categorical).alias('Borough Code'),
    pl.col('hhnd').str.lstrip().alias('House Number'),
    pl.col('bin').alias('BIN'),
    pl.col('physical_id').cast(pl.Int64, strict=False).alias('PHYSICALID'),
    pl.col('stname'),
    pl.col('zipcode'),
  ])
)
bobaadr.head(1000).tail()
```

```python
df.shape
```

```python
df = (df
  .with_column(pl.col('House Number').fill_null(''))
  .select(['House Number', 'Street Code1'])
  .join(bobaadr.select(['Street Code', 'House Number', 'BIN', 'PHYSICALID']), left_on=['Street Code1', 'House Number'], right_on=['Street Code', 'House Number'], how='left')
  .join(bobaadr.select(['Street Code', 'Borough Code', 'zipcode']).unique(subset='Street Code'), left_on=['Street Code1'], right_on=['Street Code'], how='left')
  .drop(['Street Code1', 'House Number'])
)
df.head()
```

```python
ap = pl.read_csv('data/Address_Point.csv', ignore_errors=True)
ap.head()
```

```python
ap.select(['the_geom', 'BIN', 'PHYSICALID', 'BOROCODE']).write_parquet('./data/address_point.parquet')
ap = pl.read_parquet('./data/address_point.parquet')
```

```python
ap = (ap
  .with_columns([
    pl.col('the_geom').str.slice(7).str.replace(' .*', '').cast(pl.Float64).alias('lng'),
    pl.col('the_geom').str.replace('.* ', '').str.replace('\)', '').cast(pl.Float64).alias('lat'),
  ])
  .select(['BIN', 'PHYSICALID', 'BOROCODE', 'lng', 'lat'])
)
ap.head()
```

```python
df = (df
  .join(ap, on=['BIN', 'PHYSICALID'])
  .drop(['BIN', 'PHYSICALID'])
)
```
